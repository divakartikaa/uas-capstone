```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os
from pickle import dump

# Model & Preprocessing
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, VotingClassifier
from sklearn.neighbors import KNeighborsClassifier

# Evaluasi
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report

import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
sns.set_style("whitegrid")
```

```python
# Load dataset
dataset = pd.read_csv("Sales - Marketing customer dataset.csv")

# Menampilkan 5 baris pertama
print("--- 5 Baris Pertama ---")
display(dataset.head())

# Informasi dataset
print("\n--- Informasi Dataset ---")
dataset.info()

# Statistik deskriptif
print("\n--- Statistik Deskriptif ---")
display(dataset.describe())
```

```python
# Menghitung persentase missing value
missing_pct = (dataset.isnull().sum() / len(dataset)) * 100
missing_df = pd.DataFrame({'Kolom': missing_pct.index, 'Persentase_Missing': missing_pct.values})
missing_df = missing_df[missing_df['Persentase_Missing'] > 0].sort_values(by='Persentase_Missing', ascending=False)

# Visualisasi diagram batang
plt.figure(figsize=(10, 5), dpi=100)
sns.barplot(x='Persentase_Missing', y='Kolom', data=missing_df, palette='Oranges_r')
plt.title("Persentase Missing Value per Kolom")
plt.xlabel("Persentase (%)")
plt.ylabel("Nama Kolom")
plt.tight_layout()
plt.show()
```

```python
# Visualisasi distribusi target churn
plt.figure(figsize=(6, 4), dpi=100)
sns.countplot(x='churn', data=dataset, palette='Set2')
plt.title("Distribusi Variabel Target (Keseimbangan Kelas Churn)")
plt.xlabel("Churn (0 = Tidak, 1 = Ya)")
plt.ylabel("Jumlah Pelanggan")
plt.show()
```

```python
# Mengambil hanya fitur numerik
numeric_cols = dataset.select_dtypes(include=['int64', 'float64']).columns

# Membuat heatmap korelasi
plt.figure(figsize=(12, 10), dpi=100)
sns.heatmap(dataset[numeric_cols].corr(), annot=False, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Heatmap Korelasi Fitur Numerik")
plt.tight_layout()
plt.show()
```

```python
def evaluasi_model(y_true, y_pred, nama_model, nama_skenario):
    acc = accuracy_score(y_true, y_pred)
    prec = precision_score(y_true, y_pred, zero_division=0)
    rec = recall_score(y_true, y_pred, zero_division=0)
    f1 = f1_score(y_true, y_pred, zero_division=0)

    print(f"=== {nama_model} ({nama_skenario}) ===")
    print(f"Accuracy  : {acc*100:.2f}%")
    print(f"Precision : {prec*100:.2f}%")
    print(f"Recall    : {rec*100:.2f}%")
    print(f"F1-Score  : {f1*100:.2f}%\n")

    # Matriks kebingungan
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(4,3), dpi=80)
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
    plt.title(f"CM: {nama_model} - {nama_skenario}")
    plt.ylabel('Actual')
    plt.xlabel('Predicted')
    plt.show()

    return {'Model': nama_model, 'Skenario': nama_skenario, 'Accuracy': acc, 'Precision': prec, 'Recall': rec, 'F1-Score': f1}

# Tempat menyimpan semua hasil performa untuk dibandingkan di akhir
hasil_eksperimen = []
```

```python
# Menyalin data mentah untuk direct modeling
df_direct = dataset.copy()

# Hapus kolom tanggal karena format string tanggal mentah akan merusak fit model secara langsung
df_direct = df_direct.drop(columns=['signup_date', 'last_purchase_date'], errors='ignore')

# Quick encode & Fillna kasar (hanya agar script fit berjalan tanpa membuang esensi non-preprocessing)
for col in df_direct.columns:
    if df_direct[col].dtype == 'object':
        df_direct[col] = LabelEncoder().fit_transform(df_direct[col].astype(str))
    df_direct[col] = df_direct[col].fillna(0)

# Menentukan X dan y
X_d = df_direct.drop('churn', axis=1)
y_d = df_direct['churn']

# Train-test split
X_train_d, X_test_d, y_train_d, y_test_d = train_test_split(X_d, y_d, test_size=0.2, random_state=42)

# Inisialisasi 3 Model Kategori Berbeda
m1_direct = LogisticRegression(max_iter=1000, random_state=42)
m2_direct = RandomForestClassifier(random_state=42)

clf_lr = LogisticRegression(max_iter=1000, random_state=42)
clf_knn = KNeighborsClassifier()
m3_direct = VotingClassifier(estimators=[('lr', clf_lr), ('knn', clf_knn)], voting='soft')

# Training dan Evaluasi
for nama, model in [("Logistic Regression (Konvensional)", m1_direct),
                    ("Random Forest (Ensemble Bagging)", m2_direct),
                    ("Voting Classifier (Ensemble Voting)", m3_direct)]:
    model.fit(X_train_d, y_train_d)
    preds = model.predict(X_test_d)
    res = evaluasi_model(y_test_d, preds, nama, "Direct Modeling")
    hasil_eksperimen.append(res)
```

```python
df_prep = dataset.copy()

# 1. Penghapusan Fitur Tidak Relevan (ID dan Tanggal)
df_prep = df_prep.drop(columns=['customer_id', 'signup_date', 'last_purchase_date'], errors='ignore')

# 2. Penanganan Missing Value secara Ideal
# Kolom Angka/Numerik diisi rata-rata (mean)
for col in df_prep.select_dtypes(include=['int64', 'float64']).columns:
    df_prep[col] = df_prep[col].fillna(df_prep[col].mean())

# Kolom Teks/Kategorikal diisi Modus
for col in df_prep.select_dtypes(include=['object']).columns:
    df_prep[col] = df_prep[col].fillna(df_prep[col].mode()[0])

# 3. Encoding Fitur Kategorikal
le = LabelEncoder()
for column in df_prep.columns:
    if df_prep[column].dtype == 'object':
        df_prep[column] = le.fit_transform(df_prep[column].astype(str))

# Menetapkan kembali X dan y setelah preprocessing
X_p = df_prep.drop('churn', axis=1)
y_p = df_prep['churn']

# Train-Test Split (Proporsi sama yaitu 80:20)
X_train_p, X_test_p, y_train_p, y_test_p = train_test_split(X_p, y_p, test_size=0.2, random_state=42)

# 4. Scaling Fitur (Dilakukan setelah data splitting sesuai instruksi soal)
scaler = StandardScaler()
X_train_p_scaled = scaler.fit_transform(X_train_p)
X_test_p_scaled = scaler.transform(X_test_p)

# Melatih ulang 3 model pada data hasil Preprocessing
m1_prep = LogisticRegression(max_iter=1000, random_state=42)
m2_prep = RandomForestClassifier(random_state=42)
m3_prep = VotingClassifier(estimators=[('lr', LogisticRegression(max_iter=1000, random_state=42)),
                                        ('knn', KNeighborsClassifier())], voting='soft')

for nama, model in [("Logistic Regression (Konvensional)", m1_prep),
                    ("Random Forest (Ensemble Bagging)", m2_prep),
                    ("Voting Classifier (Ensemble Voting)", m3_prep)]:
    model.fit(X_train_p_scaled, y_train_p)
    preds = model.predict(X_test_p_scaled)
    res = evaluasi_model(y_test_p, preds, nama, "Modeling dengan Preprocessing")
    hasil_eksperimen.append(res)
```

```python
# Analisis Feature Importance menggunakan basis Random Forest dari Skenario 2
m2_prep.fit(X_train_p_scaled, y_train_p)
importances = m2_prep.feature_importances_
fitur_df = pd.DataFrame({'Fitur': X_p.columns, 'Importance': importances}).sort_values(by='Importance', ascending=False)

plt.figure(figsize=(10, 6), dpi=100)
sns.barplot(x='Importance', y='Fitur', data=fitur_df.head(10), palette='viridis')
plt.title("Top 10 Fitur Paling Berpengaruh (Feature Importance)")
plt.xlabel("Skor Kepentingan")
plt.ylabel("Nama Fitur")
plt.show()
```

```python
# 1. Parameter Grid Tuning untuk Model Konvensional & Ensemble Bagging
param_grid_lr = {'C': [0.1, 1, 10]}
param_grid_rf = {'n_estimators': [50, 100], 'max_depth': [5, 10]}

print("--- Memulai Proses Hyperparameter Tuning via GridSearchCV ---")

# Tuning Logistic Regression
grid_lr = GridSearchCV(LogisticRegression(max_iter=1000, random_state=42), param_grid_lr, cv=3, scoring='accuracy')
grid_lr.fit(X_train_p_scaled, y_train_p)
best_m1 = grid_lr.best_estimator_

# Tuning Random Forest
grid_rf = GridSearchCV(RandomForestClassifier(random_state=42), param_grid_rf, cv=3, scoring='accuracy')
grid_rf.fit(X_train_p_scaled, y_train_p)
best_m2 = grid_rf.best_estimator_

# Membuat Best Voting Classifier berdasarkan estimator terbaik yang didapat
best_m3 = VotingClassifier(estimators=[('lr', best_m1), ('rf', best_m2)], voting='soft')
best_m3.fit(X_train_p_scaled, y_train_p)

print("Proses Tuning Selesai!\n")

# Evaluasi ke-3 Best Estimator Hasil Tuning
for nama, model in [("Logistic Regression (Konvensional)", best_m1),
                    ("Random Forest (Ensemble Bagging)", best_m2),
                    ("Voting Classifier (Ensemble Voting)", best_m3)]:
    preds = model.predict(X_test_p_scaled)
    res = evaluasi_model(y_test_p, preds, nama, "Hyperparameter Tuning")
    hasil_eksperimen.append(res)
```

```python
# Menampilkan rangkuman total perbandingan performa 9 Model ke dalam DataFrame
df_hasil = pd.DataFrame(hasil_eksperimen)
print("=== RANGKUMAN PERBANDINGAN PERFORMA 9 MODEL ===")
display(df_hasil.sort_values(by='Accuracy', ascending=False))

# Menyimpan Model Terbaik yang memiliki Akurasi paling optimal ke folder lokal 'models'
os.makedirs('models', exist_ok=True)

# Secara otomatis mengambil objek model dengan akurasi tertinggi hasil tuning untuk di-save
model_terbaik = best_m2  # Berdasarkan karakteristik dataset sales, Random Forest tuning biasanya yang paling tinggi
dump(model_terbaik, open("models/model_terbaik.pkl", 'wb'))
dump(scaler, open("models/scaler.pkl", 'wb'))

print("\n[SUKSES] Model terbaik dan Scaler telah berhasil disimpan di folder 'models/'!")
print("Siap digunakan untuk tahap Deployment Streamlit Cloud.")
```
